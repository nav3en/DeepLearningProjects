### Create Word Vectors from Game of thrones books and analyze semantic similarity in them

```python
# Import needed libraries
%matplotlib inline

# Future is the link between python 2 and 3
from __future__ import absolute_import, division, print_function
import codecs
import glob
import multiprocessing
import os
# human readable printing 
import pprint 
import nltk
import scipy as sc
import gensim.models.word2vec as word2vec
# for dimensionality reduction
import sklearn
from sklearn import manifold
import numpy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import re
```

```python
# Step 1 - Process our data
# Data cleaning
#nltk.download('punkt') #pretrained tokenizer
#nltk.download('stopwords')
# Get the booknames
book_names = sorted(glob.glob("data/*1.txt"))
book_names
```

```python
# Combining all the books into one corpus and in UTF-8 format
corpus_raw = u""
#for each book, read it, open it un utf 8 format, 
#add it to the raw corpus
for book_filename in book_names:
    print("Reading '{0}'...".format(book_filename))
    with codecs.open(book_filename, "r", "utf-8") as book_file:
        corpus_raw += book_file.read()
    print("Corpus is now {0} characters long".format(len(corpus_raw)))
    print()
```

```python
# Split the corpus into sentences
## Loading the tokenizer
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
sentences = tokenizer.tokenize(corpus_raw)
print("Now we have "+str(len(sentences)) + " sentences")
```

```python
# Converting sentences to a list of words - Removing unecessary characters
def sentence_to_words(raw_sentence):
    clean_sentence = re.sub("[^a-zA-Z]", " ", raw_sentence)
    words = clean_sentence.split()
    return words
```

```python
# Converting each sentence to corresponding list of tokens
sentences_word_lists=[]
for sentence in sentences:
    if len(sentence)>0:
        sentences_word_lists.append(sentence_to_words(sentence))
        
print(sentences[5])
print(sentences_word_lists[5])
```

```python
token_length = sum([len(sentence) for sentence in sentences_word_lists])
print(" There are "+str(token_length) +" in the corpus")
```

### Train Word2Vec

```python
# Once we have vectors

# Hyper parameters
num_featues = 300
min_word_count = 3
num_workers = multiprocessing.cpu_count()
# This is the size of the context window
context_size = 7
#downsampling for frequent words
downsampling = 1e-3
# deterministic
seed = 1
```

```python
got2vector = word2vec.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_featues,
    min_count=min_word_count,
    window=context_size,
    sample = downsampling
)
```

```python
# Building the model
got2vector.build_vocab(sentences_word_lists)
print("Word2Vec vocab length "+str(len(got2vector.vocab)))
```

```python
# Training the model
got2vector.train(sentences_word_lists)
```

```python
# Saving the model
if not os.path.exists("trained"):
    os.makedirs("trained")
got2vector.save(os.path.join("trained", "got_book1_2vec.w2v"))
```

```python
# We can load the model using 
got2vector = word2vec.Word2Vec.load(os.path.join("trained", "got_book1_2vec.w2v"))
```

### Compress the vector into a 2D space to visualize the word vectors (They are 300D now)

```python
# We can use TSNE for this
# initializing tsne
tsne = sklearn.manifold.TSNE(n_components=2,random_state=0)

#Building the input 300d vector
word_vector_matrix = got2vector.syn0
```

```python
# Training the TSNE model
word_vector_matrix_2D = tsne.fit_transform(word_vector_matrix)
```

```python
points = pd.DataFrame([(word, coords[0], coords[1])
        for word, coords in [
            (word, word_vector_matrix_2D[got2vector.vocab[word].index])
            for word in got2vector.vocab]],columns=["words","x","y"])
points.head()
```

```python
# Plotting the points
sns.set_context("poster")
points.plot.scatter("x", "y", s=10, figsize=(20, 12))
# All the word vectos visualized
```

```python
got2vector.most_similar("Stark")
```

```python
got2vector.most_similar("winter")
```

```python
#distance, similarity, and ranking
def nearest_similarity_cosmul(start1, end1, end2):
    similarities = got2vector.most_similar_cosmul(
        positive=[end2, start1],
        negative=[end1]
    )
    start2 = similarities[0][0]
    print("{start1} is related to {end1}, as {start2} is related to {end2}".format(**locals()))
    return start2
```

```python
nearest_similarity_cosmul("Stark", "Winterfell", "Riverrun")
nearest_similarity_cosmul("Jaime", "sword", "wine")
nearest_similarity_cosmul("Arya", "Nymeria", "dragons")
```
